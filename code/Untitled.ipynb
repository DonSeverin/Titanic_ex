# Assignment I: ODI Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Getting CSV data loaded
df = pd.read_csv("train.csv")
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df.head()
```

```python
df.describe()
```

```python
df.describe(include=['O'])
```

```python
g = sns.FacetGrid(df, col="Sex", hue="Survived")
g.map(sns.histplot, "Pclass")
g.add_legend()
plt.savefig("Vizualizations/PclassSurvival.png")
```

```python
g = sns.FacetGrid(df, col="Parch", hue="Survived")
g.map(sns.histplot, "Age")
g.add_legend()
plt.savefig("Vizualizations/ParchageSurvival.png")
```

```python
g = sns.FacetGrid(df, col="SibSp", hue="Survived")
g.map(sns.histplot, "Age")
g.add_legend()
plt.savefig("Vizualizations/SibSpageSurvival.png")
```

```python
g = sns.FacetGrid(df, col="Parch", col_wrap=4, height=2, ylim=(0, 10))
g.map(sns.pointplot, "Age", "Survived", order=[1, 2, 3], color=".3", ci=None)
```
